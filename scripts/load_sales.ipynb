```python
# Base path to csv files
base_path = "../raw-files/Sales Data/"

# List file names
file_names = [
    'Sales CountryRegionCurrency.csv',
    'Sales CreditCard.csv',
    'Sales Currency.csv',
    'Sales CurrencyRate.csv',
    'Sales Customer.csv',
    'Sales PersonCreditCard.csv',
    'Sales SalesOrderDetail.csv',
    'Sales SalesOrderHeader.csv',
    'Sales SalesOrderHeaderSalesReason.csv',
    'Sales SalesPerson.csv',
    'Sales SalesPersonQuotaHistory.csv',
    'Sales SalesReason.csv',
    'Sales SalesTaxRate.csv',
    'Sales SalesTerritory.csv',
    'Sales SalesTerritoryHistory.csv',
    'Sales ShoppingCartItem.csv',
    'Sales SpecialOffer.csv',
    'Sales SpecialOfferProduct.csv',
    'Sales Store.csv'
]
```

```python
from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder.appName("adventure-works").getOrCreate()

# Read each file into a DataFrame
dataframes = {}
for file_name in file_names:
    df_name = file_name.replace(' ', '_').replace('.csv', '').lower()
    dataframes[df_name] = spark.read.format('csv').option('header', 'true').load(f'{base_path}/{file_name}')

# Accessing individual DataFrames with the correct keys
sales_countryregioncurrency_df = dataframes['sales_countryregioncurrency']
sales_creditcard_df = dataframes['sales_creditcard']
sales_currency_df = dataframes['sales_currency']
sales_currencyrate_df = dataframes['sales_currencyrate']
sales_customer_df = dataframes['sales_customer']
sales_personcreditcard_df = dataframes['sales_personcreditcard']
sales_salesorderdetail_df = dataframes['sales_salesorderdetail']
sales_salesorderheader_df = dataframes['sales_salesorderheader']
sales_salesorderheadersalesreason_df = dataframes['sales_salesorderheadersalesreason']
sales_salesperson_df = dataframes['sales_salesperson']
sales_salespersonquotahistory_df = dataframes['sales_salespersonquotahistory']
sales_salesreason_df = dataframes['sales_salesreason']
sales_salestaxrate_df = dataframes['sales_salestaxrate']
sales_salesterritory_df = dataframes['sales_salesterritory']
sales_salesterritoryhistory_df = dataframes['sales_salesterritoryhistory']
sales_shoppingcartitem_df = dataframes['sales_shoppingcartitem']
sales_specialoffer_df = dataframes['sales_specialoffer']
sales_specialofferproduct_df = dataframes['sales_specialofferproduct']
sales_store_df = dataframes['sales_store']
```

```python
def rename_columns(df, rename_mapping):
    for old_name, new_name in rename_mapping.items():
        df = df.withColumnRenamed(old_name, new_name)
    return df

# Define the rename mappings for each DataFrame
# Define the rename mappings for each DataFrame
rename_mappings = {
    "sales_countryregioncurrency_df": {"ModifiedDate": "SalesCountryRegionModifiedDate"},
    "sales_creditcard_df": {"ModifiedDate": "SalesCreditCardModifiedDate"},
    "sales_currency_df": {"ModifiedDate": "SalesCurrencyModifiedDate", "Name": "SalesCurrencyName"},
    "sales_currencyrate_df": {"ModifiedDate": "SalesCurrencyRateModifiedDate"},
    "sales_customer_df": {"ModifiedDate": "SalesCustomerModifiedDate", "rowguid": "SalesCustomerrowguid", "TerritoryID":"SalesCustomerTerritoryID", "AccountNumber":"SalesCustomerAccountNumber", "AccountName": "SalesCustomerAccountName"},
    "sales_personcreditcard_df": {"ModifiedDate": "SalesPersonCreditCardModifiedDate"},
    "sales_salesorderdetail_df": {"ModifiedDate": "SalesOrderDetailModifiedDate", "rowguid": "SalesOrderDetailrowguid"},
    "sales_salesorderheader_df": {"ModifiedDate": "SalesOrderHeaderModifiedDate", "rowguid": "SalesOrderHeaderrowguid", "AccountNumber":"SalesSalesOrderHeaderAccountNumber"},
    "sales_salesorderheadersalesreason_df": {"ModifiedDate": "SalesOrderHeaderSalesReasonModifiedDate"},
    "sales_salesperson_df": {"ModifiedDate": "SalesPersonModifiedDate", "rowguid": "SalesPersonrowguid", "TerritoryID":"SalesSalePersonTerritoryID", "BusinessEntityID":"SalesSalesPersonBusinessEntityID", "AccountName": "SalesPersonAccountName"},
    "sales_salespersonquotahistory_df": {"ModifiedDate": "SalesPersonQuotaHistoryModifiedDate", "rowguid": "SalesPersonQuotaHistoryrowguid", "AccountName": "SalesPersonQuotaHistoryAccountName"},
    "sales_salesreason_df": {"ModifiedDate": "SalesReasonModifiedDate", "Name": "SalesReasonName"},
    "sales_salestaxrate_df": {"ModifiedDate": "SalesTaxRateModifiedDate", "rowguid": "SalesTaxRaterowguid", "Name": "SalesTaxRateName"},
    "sales_salesterritory_df": {"ModifiedDate": "SalesTerritoryModifiedDate", "rowguid": "SalesTerritoryrowguid"},
    "sales_salesterritoryhistory_df": {"ModifiedDate": "SalesTerritoryHistoryModifiedDate", "TerritoryID":"SalesSalesTerritoryHistoryTerritoryID", "AccountName": "SalesTerritoryHistoryAccountName"},
    "sales_shoppingcartitem_df": {"ModifiedDate": "ShoppingCartItemModifiedDate", "ProductID":"SalesShoppingCartItemProductID"},
    "sales_specialoffer_df": {"ModifiedDate": "SpecialOfferModifiedDate", "rowguid": "SpecialOfferrowguid"},
    "sales_specialofferproduct_df": {"ModifiedDate": "SpecialOfferProductModifiedDate", "rowguid": "SpecialOfferProductrowguid", "ProductID":"SalesSpecialOfferProductProductID"},
    "sales_store_df": {"ModifiedDate": "StoreModifiedDate", "rowguid": "Storerowguid", "Name": "StoreName", "SalesPersonID": "SalesStoreSalesPersonID", "AccountName": "SalesStoreAccountName"},
}
```

```python
# Apply the rename mappings to the DataFrames
sales_countryregioncurrency_df = rename_columns(sales_countryregioncurrency_df, rename_mappings["sales_countryregioncurrency_df"])
sales_creditcard_df = rename_columns(sales_creditcard_df, rename_mappings["sales_creditcard_df"])
sales_currency_df = rename_columns(sales_currency_df, rename_mappings["sales_currency_df"])
sales_currencyrate_df = rename_columns(sales_currencyrate_df, rename_mappings["sales_currencyrate_df"])
sales_customer_df = rename_columns(sales_customer_df, rename_mappings["sales_customer_df"])
sales_personcreditcard_df = rename_columns(sales_personcreditcard_df, rename_mappings["sales_personcreditcard_df"])
sales_salesorderdetail_df = rename_columns(sales_salesorderdetail_df, rename_mappings["sales_salesorderdetail_df"])
sales_salesorderheader_df = rename_columns(sales_salesorderheader_df, rename_mappings["sales_salesorderheader_df"])
sales_salesorderheadersalesreason_df = rename_columns(sales_salesorderheadersalesreason_df, rename_mappings["sales_salesorderheadersalesreason_df"])
sales_salesperson_df = rename_columns(sales_salesperson_df, rename_mappings["sales_salesperson_df"])
sales_salespersonquotahistory_df = rename_columns(sales_salespersonquotahistory_df, rename_mappings["sales_salespersonquotahistory_df"])
sales_salesreason_df = rename_columns(sales_salesreason_df, rename_mappings["sales_salesreason_df"])
sales_salestaxrate_df = rename_columns(sales_salestaxrate_df, rename_mappings["sales_salestaxrate_df"])
sales_salesterritory_df = rename_columns(sales_salesterritory_df, rename_mappings["sales_salesterritory_df"])
sales_salesterritoryhistory_df = rename_columns(sales_salesterritoryhistory_df, rename_mappings["sales_salesterritoryhistory_df"])
sales_shoppingcartitem_df = rename_columns(sales_shoppingcartitem_df, rename_mappings["sales_shoppingcartitem_df"])
sales_specialoffer_df = rename_columns(sales_specialoffer_df, rename_mappings["sales_specialoffer_df"])
sales_specialofferproduct_df = rename_columns(sales_specialofferproduct_df, rename_mappings["sales_specialofferproduct_df"])
sales_store_df = rename_columns(sales_store_df, rename_mappings["sales_store_df"])
```

```python
# Initialize the DataFrame with a join between sales_order_header and sales_customer
sales_details_df = sales_salesorderheader_df.join(sales_customer_df, 'CustomerID', 'left')

# Join with SalesStore using the StoreID
sales_details_df = sales_details_df.join(sales_store_df, sales_details_df.StoreID == sales_store_df.BusinessEntityID, 'left')

# Join with SalesTerritory using the TerritoryID
sales_details_df = sales_details_df.join(sales_salesterritory_df, 'TerritoryID', 'left')


# Join with SalesPerson using the SalesPersonID
sales_details_df = sales_details_df.join(sales_salesperson_df, sales_details_df.SalesPersonID == sales_salesperson_df.SalesSalesPersonBusinessEntityID, 'left')

# Join with SalesTerritoryHistory using the TerritoryID
sales_details_df = sales_details_df.join(sales_salesterritoryhistory_df, sales_details_df.TerritoryID==sales_salesterritoryhistory_df.SalesSalesTerritoryHistoryTerritoryID, 'left')

# Join with SalesOrderDetail using the SalesOrderID
sales_details_df = sales_details_df.join(sales_salesorderdetail_df, 'SalesOrderID', 'left')

# Join with ShoppingCartItem using the ProductID
sales_details_df = sales_details_df.join(sales_shoppingcartitem_df, sales_details_df.ProductID == sales_shoppingcartitem_df.SalesShoppingCartItemProductID, 'left')

# Join with SpecialOfferProduct using the ProductID
sales_details_df = sales_details_df.join(sales_specialofferproduct_df, sales_details_df.ProductID == sales_specialofferproduct_df.SalesSpecialOfferProductProductID, 'left')

# Join with SpecialOffer using the SpecialOfferID
sales_details_df = sales_details_df.join(sales_specialoffer_df, sales_specialofferproduct_df.SpecialOfferID == sales_specialoffer_df.SpecialOfferID, 'left')

# Join with SalesOrderHeaderSalesReason using the SalesOrderID
sales_details_df = sales_details_df.join(sales_salesorderheadersalesreason_df, 'SalesOrderID', 'left')

# Join with SalesReason using the SalesReasonID
sales_details_df = sales_details_df.join(sales_salesreason_df, sales_salesorderheadersalesreason_df.SalesReasonID == sales_salesreason_df.SalesReasonID, 'left')

# Join with SalesCreditCard using the CreditCardID
sales_details_df = sales_details_df.join(sales_creditcard_df, 'CreditCardID', 'left')

# Join with PersonCreditCard using the CreditCardID
sales_details_df = sales_details_df.join(sales_personcreditcard_df, sales_details_df.CreditCardID == sales_personcreditcard_df.CreditCardID, 'left')

# Join with SalesCurrency using the CurrencyCode
# sales_details_df = sales_details_df.join(sales_currency_df, sales_details_df.CurrencyCode == sales_currency_df.CurrencyCode, 'left')

# Join with SalesCountryRegionCurrency using the CurrencyCode
# sales_details_df = sales_details_df.join(sales_countryregioncurrency_df, sales_details_df.CurrencyCode == sales_countryregioncurrency_df.CurrencyCode, 'left')

# Join with SalesCurrencyRate using the CurrencyRateID
# sales_details_df = sales_details_df.join(sales_currencyrate_df, 'CurrencyRateID', 'left')

# Join with SalesTaxRate using StateProvinceID matching to TerritoryID
sales_details_df = sales_details_df.join(sales_salestaxrate_df, sales_details_df.TerritoryID == sales_salestaxrate_df.StateProvinceID, 'left')

sales_details_df = sales_details_df.drop('BusinessEntityID')
sales_details_df = sales_details_df.drop('creditcardid')
sales_details_df = sales_details_df.drop('enddate')
sales_details_df = sales_details_df.drop('startdate')
sales_details_df = sales_details_df.drop('saleslastyear')
sales_details_df = sales_details_df.drop('salesreasonid')
sales_details_df = sales_details_df.drop('salesytd')
sales_details_df = sales_details_df.drop('specialofferid')
```
